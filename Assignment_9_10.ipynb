```python
import pandas as pd

movies = pd.read_csv (r'movies.csv')
```

```python
movies.columns
```

```python
movies.head()
```

```python
movies_color = movies.corr()
movies_color.style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```

Alle correlaties tussen de numerieke kolommen van de movies dataset.

----

Correlatie zien tussen het aantal votes en revenue.

```python
movies[['vote_count','revenue']].corr()
```

```python
movies_color.plot(kind='scatter', x='vote_count', y='revenue', s=30)
```

Positive correlation: 0.8

Een hoog budget betekent meestal dat het een bekende film is. Bekende films worden vaker beoordeeld.

----

Correlatie zien tussen het budget en de opbrengst van de films.

```python
movies[['budget','revenue']].corr()
```

```python
movies_color.plot(kind='scatter', x='revenue', y='budget', s=30)
```

Positive correlation: 0.7

```python
# Excluding a couple of columns because their values are too high to show in the same plot
movies.drop(['id', 'vote_count', 'runtime', 'popularity', 'vote_average'], axis=1).plot(figsize=(18,5))
```

Het resultaat is zoals verwacht. Meestal hoe hoger het budget, hoe hoger de revenue. Er zijn enkele uitschieters waarbij het budget hoger is dan de revenue (verlies).

----

Correlatie zien tussen de runtime en populariteit van de films.

```python
movies[['popularity','runtime']].corr()
```

```python
movies_color.plot(kind='scatter', x='runtime', y='popularity', s=30)
```

Positive correlation: 0.2

```python
# Excluding a couple of columns because their values are too high to show in the same plot
movies.drop(['id', 'vote_count', 'revenue', 'budget', 'vote_average'], axis=1).plot(figsize=(18,5))
```

Hier zie je dat er een hele kleine samenhang ligt tussen de populariteit van een film en de afspeelduur. Films die meer an 3 uur lang zijn, zijn meestal niet al te populair. Evenals de 'short films'.
